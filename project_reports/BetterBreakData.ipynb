```python
import pandas as pd
import numpy as np
```

```python
b1 = pd.read_csv("Main_Breaks_Post_March_2022.csv")
b2 = pd.read_csv("Main_Breaks_Jan_2013_to_March_2022.csv")
b3 = pd.read_csv("Main_Breaks_Pre_2013.csv")
```

```python
b1_clean = b1.rename(columns={
    "entityuid": "pipe_tag",
    "initiateda": "break_date"
})[["pipe_tag", "break_date"]]

b2_clean = b2.rename(columns={
    "pipe_tag": "pipe_tag",
    "date_repor": "break_date"
})[["pipe_tag", "break_date"]]

b3_clean = b3.rename(columns={
    "tag": "pipe_tag",
    "wodate": "break_date"
})[["pipe_tag", "break_date"]]
```

```python
b1_clean["break_date"] = pd.to_datetime(b1_clean["break_date"], errors="coerce")
b2_clean["break_date"] = pd.to_datetime(b2_clean["break_date"], errors="coerce")
b3_clean["break_date"] = pd.to_datetime(b3_clean["break_date"], errors="coerce")
```

```python
b1_clean.isna().sum()
```

```python
b2_clean.isna().sum()
```

```python
b3_clean.isna().sum()
```

```python
b3_clean = b3_clean.dropna()
```

```python
b3_clean.isna().sum()
```

```python
breaks1 = pd.concat([b1_clean, b2_clean], ignore_index=True)
```

```python
breaks2 = pd.concat([b1_clean, b2_clean, b3_clean], ignore_index=True)
```

```python
breaks1.head()
```

```python
breaks2.tail()
```

```python
breaks1.to_csv("Breaks2013toNow.csv", index=False)
```

```python
breaks2.to_csv("BreaksWithHistorical.csv", index=False)
```
